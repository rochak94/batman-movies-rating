**Data Visualization**

Imdb Ratings visualization 

```python
def find_minimax(x):
  min_index=batman[x].idxmin()
  high_index=batman[x].idxmax()
  highest=pd.DataFrame(batman.iloc[high_index])
  lowest=pd.DataFrame(batman.iloc[min_index])
  print('Batman Movie with lowest Rating' + x + ':', batman['Title'][min_index])
  print('Batman Movie with highest Rating' + x + ':', batman['Title'][high_index])
  return pd.concat([lowest,highest], axis=1)  

find_minimax('Imdb Rating')
```

```python
detail=pd.DataFrame(batman['Imdb Rating'].sort_values(ascending=False))
detail['Title'] = batman['Title']
data = list(map(str,detail['Title']))

x = list(data[:10])
y = list(detail['Imdb Rating'] [:10])

ax = sns.barplot(x=x , y=y)

sns.set(rc={'figure.figsize' : (20,10)})
ax.set_title('Top 10 titles', fontsize = 30)
ax.set_xlabel('Imdb Rating', fontsize=20)
plt.xticks(rotation = 90)
```

Movies released in the years

```python
Rel_Year = batman.groupby('Year').count()['Title']
print(Rel_Year.head())
```

```python
batman.groupby('Year').count()['Title'].plot()

sns.set(rc={'figure.figsize': (20,10)})
plt.title('Year vs. number of Movies', fontsize = 30)
plt.xlabel('Release year', fontsize=20)
plt.ylabel('Number of Titles', fontsize = 20)
sns.set_style('darkgrid')
```

```python
batman_dir= batman.groupby('Year').count()['Director']
print(batman_dir.head())
```

```python
batman.groupby('Director').count()['Year'].plot()

sns.set(rc={'figure.figsize': (20,10)})
plt.title('Year vs. Director', fontsize = 30)
plt.xlabel('Director', fontsize=20)
plt.ylabel('Amount of Titles', fontsize = 20)
plt.xticks(rotation = 90)
sns.set_style('darkgrid')
```
